Importing Libraries

```python
import cv2
from PIL import Image
from IPython.display import display
```

Importing Image 

```python
image = cv2.imread('/content/drive/MyDrive/DATASETS/nft.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

Turning Color Image to Black and White

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

Turning Black and White photo to Negative Image

```python
inverted_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(inverted_image))
```

Converting Negative Image to Blurry Image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
display(Image.fromarray(blurred_image))
```

Image in Pencil Sketch Form

```python
inverted_blurred = cv2.bitwise_not(blurred_image)
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=250.0)
display(Image.fromarray(pencil_sketch))
```
